```python
import numpy as np
import matplotlib.pyplot as plt
import typing

from gaussian_math import GaussianVariable

def plot_pdf(var : GaussianVariable, ax):
        x = np.linspace(
            *ax.get_xlim(),
            1000
        )
        y = var.pdf(x)
        ax.plot(x, y)
```

```python
X = GaussianVariable(5, 2)
Y = GaussianVariable(6, 8)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.hist(X.sample() + Y.sample(), density=True, bins=100)
ax.set_title("{} + {}".format(X, Y))
plot_pdf(X + Y, ax)
```

```python
X = GaussianVariable(5, 2)
Y = GaussianVariable(6, 8)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.hist(X.sample() - Y.sample(), density=True, bins=100)
ax.set_title("{} - {}".format(X, Y))
plot_pdf(X - Y, ax)
```
